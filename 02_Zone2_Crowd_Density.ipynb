```python
from shapely.geometry import Point, Polygon
import folium
```

```python
# Zone 2 Extreme Points - Specified in this array of (long,lat),(long,lat)...
coordinates = [(-43.24587, -22.81541), (-43.24592, -22.81525), (-43.24585, -22.81511), (-43.24566, -22.81504), (-43.24545, -22.81508), 
               (-43.24533, -22.81527), (-43.24534, -22.81544), (-43.24554, -22.81555), (-43.24575, -22.81552)]
```

```python
#Function to create Polygon (in this case Zone 2 - kind of a circle)
def create_polygon(coords, polygon_name):
 polygon = Polygon(coordinates)
 gdf = gpd.GeoDataFrame(crs = {"init" :"epsg:4326"})
 gdf.loc[0,"name"] = polygon_name
 gdf.loc[0, "geometry"] = polygon
 return gdf
```

```python
zone2 = create_polygon(coordinates,"zone2")
```

```python
# Functions to tell if point is inside or outside
mask = (zone2.loc[0, "geometry"])
pip_mask_geofence = gdf.within(mask)
```

```python
#Create PIP mask
gdf.loc[:,"geofence"] = pip_mask_geofence
```

```python
gdf['geofence'] = gdf['geofence'].replace({True: 'Inside', False: 'Outside'})
```

# 2° Step - Crowd Density on Zone 2

```python
# Count how many Points are inside the Zone2
a = gdf[gdf['geofence'] == "Inside"]
points_inside = a.shape[0]
points_inside
```

```python
inside = gdf[gdf['geofence'] == "Inside"]
```

```python
#Count how many unique devices are there in the zone
uniq_mac = list(inside['ClientMacAddr'].unique())
devices_inside = len(uniq_mac)
devices_inside
```

```python
#Set up a customizable threshold per zone and per hour maybe
threshold = 200
```

```python
if devices_inside > threshold:
  alarm = "Warning: too many devices in Zone 2. Crowd Density above allowed threshold"
if devices_inside <= threshold:
  alarm = "Situation OK: Crowd Density in Zone 2 under control"
```
